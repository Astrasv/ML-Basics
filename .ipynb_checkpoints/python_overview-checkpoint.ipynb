```python
store_sales = pd.read_csv("store_sales.csv")
```

```python
store_sales.head()
```

```python
# find out avg sales for each month
store_sales.apply("mean", axis = 0) #axis = 0 is default
store_sales.select_dtypes(include  = ['int64']).apply("mean", axis = 0)
```

```python
# find out avg sales for each store
store_sales.apply("mean", axis = 1)
```

```python
store_sales.groupby("city").agg('mean')
```

```python
def bonus_func(sales_numbers):
    #List comprehension for the below code --> return ["Eligible" if x > 15 else "Not Eligible"  for x in sales_numbers]
    ls = []
    for x in sales_numbers:
        if x>15:
            ls.append("Eligible")
        else:
            ls.append("Not Eligible")
    return ls

def bonus_funcnp(sales_col):
    #Using np.where() method instead of list comprehension
    elig_vals = np.where(sales_col>10, "Eligible","Not Eligible")
    return elig_vals
```

```python
newdf = store_sales.select_dtypes(include = ['int64','float64']).apply(bonus_funcnp)
newdf
```

```python
newdf1 = store_sales.select_dtypes(['int64','float64']).apply(bonus_func)
newdf1
```

```python
final = pd.concat([store_sales,newdf], axis = 1)       #Concatinates the dataframes along the given axis
final
```

```python
bonus_func([12,23,56])
```

```python
store_sales.iloc[:,2:].apply(bonus_func).head()
```

```python
store_sales.iloc[:,2:].apply(lambda sales_numbers: ["Eligible" if x > 15 else "Not Eligible" for x in sales_numbers]).head()
```

### Matrix operations use case

### Example 1 - compute month on month growth in sales

```python
store_sales.head()
```

```python
sales_mat = np.array(store_sales.iloc[:,2:])
```

```python
sales_mat
```

```python
sales_mat_lag = sales_mat[:,1:]
sales_mat_lag
```

```python
sales_mat_lag = sales_mat[:,1:]
sales_mat_lag = np.c_[sales_mat_lag,np.zeros(100)]
```

```python
sales_diff = sales_mat_lag - sales_mat
```

```python
perct_change = pd.DataFrame(sales_diff/sales_mat)
```
